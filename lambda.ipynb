```python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(6))
```

```python
# find the n binary numbers for decimal nuber using recursion
def binary(n):
    if n > 1:
        binary(n // 2)
    print(n % 2, end="")

n = int(input("Enter a decimal number: "))
binary(n)
```

```python
# sum of n natural numbers by using recursion
def sum_natural(n):
    if n == 0:
        return 0
    return n + sum_natural(n - 1)

n = int(input())
print("Sum =", sum_natural(n))
```

```python
def product(n):
    if n == 0:
        return 0
    return n + product(n - 1)

n = int(input())
print("product =", product(n))
```

```python
def decimal_to_binary(n):
    if n == 0:
        return ""
    return decimal_to_binary(n // 2) + str(n % 2)

n = int(input())

if n == 0:
    print("Binary:", 0)
else:
    print("Binary:", decimal_to_binary(n))
```

```python
def binary(n):
    if n > 1:
        binary(n // 2)
    print(n % 2, end="")

n = int(input())

if n == 0:
    print(0)
else:
    binary(n)
```

```python
# Sum of two numbers using lambda

add = lambda a, b: a + b

num1 = int(input())
num2 = int(input())

print("Sum =", add(num1, num2))
```

```python
# Lambda function to calculate squares
square = lambda x: x * x

num1 = int(input())
num2 = int(input())

print("Square of", num1, "=", square(num1))
print("Square of", num2, "=", square(num2))
```

```python
# find square of all list of integer using lambda and map()
arr =[1,2,3,4,5]
res = list(map(lambda x: x**2,arr))
print(res)
```

```python
#find the even numbers by using filter and lambda
arr =[1,2,3,4,5]
res = list(filter(lambda x: x%2==0,arr))
print(res)
```

```python
#single line if-else syntax
#True_exp if condition else False_exp
#find the wheather the number is even or odd by using lambda
n = int(input())
even_or_odd = lambda x:"even" if x%2 else "odd"
print(even_or_odd(n))
```

```python
n = int(input())
even_or_odd = lambda x:f"{x}-odd" if x%2 else f"{x}-even"
print(even_or_odd(n))
```
