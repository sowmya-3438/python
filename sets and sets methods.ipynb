### Set Methods
-Add Element To Set
-add (ele):It adds element to set if element not exists in set ,otherwise nothing 
- update(iterable): it adds all iterable to set
-Remove elements from set
-remove(ele):it remove the element from set if element exists in set otherwise'keyerror'
-discard(ele): it removes the element from set if element exists in set otherwise nothing
-Set Operations
-pop():it remove and returns the element from set
-union(anthor_sets):It returns the all common elements from all sets
-intersection(anthor_sets):it returns the common elements from the all sets
- difference(anthor_sets):It returns the set elements those are not lies in another_set
- symmetric_difference(anthor_sets):It returns the Non-common elements from all sets



```python
s = {1,2,3,4, 'hello','sowmya'}
s.add(6)
print(s)
s.remove(4)
print(s)
s.discard('hello')
print(s)
s.discard('code')
print(s)
s.update('sow')
print(s)
s.pop('hi')
print(s)
```

```python
s1 ={1,4,5,6,7,7,8}
s2 ={2,4,2,5,6,7,8}
s3 = {3,4,4,5,5,6}
print(s1.union(s2,s3))
print(s2.intersection(s2,s1))
print(s2.intersection(s3,s1))
print(s3.intersection(s1))
print(s2.difference(s3,s1))
print(s2.difference(s1))
print(s3.difference(s1))
```

```python
s1 ={1,4,5,6,7,7,8}
s2 ={2,4,2,5,6,7,8}
s3 = {3,4,4,5,5,6}
print(s1 | s2) #|-union
print(s1 | s2 | s3)
print(s2 & s1) # & -intersection
print(s2 & s3 & s1)
print(s3 & s1)
print(s2 - s1) # - for difference
print(s3 - s1)
print(s3 - s1)
```

```python
s1 ={1,2,3,3}
s2 ={3,4,5}
s1.intersection_update(s2)
print(s1)
s1.difference_update(s2)
print(s1,s2)
s1.symmetric_difference_update(s2)
print(s1)
```
