```python
#abs():it returns the abuslute value
print(abs(20-10))
print(abs(10-20))
```

```python
#ord(char):it returns ASCII value of chsrscter
#chr(ascii_val):it returns the character for specified ASCII value
print(ord("A"))
print(chr(65))
```

```python
#sorted(iterable, reverse=False,keynone):
#sort the fruits based on the length of fruit name
fruits = ['Mango','Apple','bananan','kivi','Water melon']
res = sorted(fruits)
print(res)
print(sorted(fruits,reverse=True))
print(sorted(fruits,reverse=True,key = len))# it sorts the list based on length
```

```python
#map(function,itersble)
#find the square of all list of integer using map funtion assign 
def square(x):
    return x**2
arr = [1,2,3,4,5]
res = list(map(square,arr))
print(res)
```

```python
#filter(function,iterable):it filters the iterable if the function returns non-none those values will added to result
#find even numbers from list of numbers
def check_even(x):
    if x % 2 == 0:
        return x
lst = [1,2,3,4,5,6]
res=list(filter(check_even,lst))
print(res)
```

```python
#find the 3's multipliers from list of integer
arr = [1,24,54,22,44,51,33,15]
def Threes_multipliers(x):
    if x%3==0:
        return 1
res = list(filter(Threes_multipliers, arr))
print(res)
```

```python
#enumerate(iterable):This function returns the list of tuple of index and item for iterable items
fruits = ['Mango','Apple','bananan','kivi','Water melon']
print(enumerate(fruits))
print(list(enumerate(fruits)))
```

```python
#zip(iterable1,iterable2,..):it returns all iterable as single iterable of tuples
rool_no = [101,102,103,104,105]
names = ['Ravi','Raju','siva','Rani','Gopi']
percentage =[96,67,91.8,67.65,99.9999]
print(list(zip(rool_no, names, percentage)))
```

```python
percentage = [34,35,54,65,67]
for num in percentage:
    if num < 35:
        print("All Students not passed")
        break
else:
    print("All students pass")
```

```python
percentage = [34,35,54,65,67]
lst = []
for num in percentage:
    if num < 35:
      lst.append(False)
    else:
        lst.append(True)
print(lst)
if all(lst):
    print("All students pass")
else:
    print("All Students not passed")
```
