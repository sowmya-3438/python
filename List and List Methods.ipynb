####List operations
1.List concatenation(using '+' operator)
2.List reperation(using '*' operator)

```python
#List concatenation
l1 = [1,2,3,4]
l2 = [4,5,6]
res = l1 + l2
print(res)
print(l1 + l2)
# List reperation
print(l1 * 3)
print([0] * 10)
print(10 * [1])
print([] * 10)
```

```python
### List Methods
-Element Adding Methods
-apppend(item):It adds item at end at the list
-extand(itrable): It adds all iterable elements at end of the list
-insert(pos, item):This method adds the item at specified position
```

```python
lst = [1,2,3,4]
print(lst.append(10))
print(lst)
lst.append('code')
print(lst)
lst.append([10,20,30])
print(lst)
print(lst[6][1])
lst.extend([100,200,300])
print(lst)
lst.insert(5,2000)
print(lst)
```

```python
-Element Remove Method
-remove(val):It removes the first occurance of value if value exists in list,otherwise it raises 'ValueError'
-pop():It removes  and returns the last of element of list
-pop(pos):It removes and returns the specified position element from list
-clear():It removes all element from list
-del:It python,keyword,used to delete values
```

```python
lst = [10,20,30,20,40]
lst.remove(20)
print(lst)
#lst.remove(100)# ValueError :list.remove(x):x is not list
val = lst.pop()
print(val,lst)
val2 = lst.pop(1)# it removes the 1st position element
print(val2,lst)
#print(lst.pop(10))
del lst[0]
print(lst)
lst.clear()
print(lst)
```

```python
## index (val): It returns the index of first occurance of value if value exists in list otherwise value error
## count(val):it returns the total frequence of value
lst = [1,2,34,45,56,7,8,9]
ind = lst.index(2)
print(ind)
print(lst.count(2))
print(lst.count(100))
print(lst.index(100))
```
